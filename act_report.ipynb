A visualization report about the weratedogs wrangle project which was gathered from three datasets and assessed and cleaned into one dataframe.

  - timestamp: Day the tweet was tweeeted

  - Text: the tet written in the tweet
  - name : the name of the dogs 
  - dog_stage : the stage of the dog there are 4 main stage(doggo,puppo,puper,flooper)
  - rating_denominator : the rating at the top
  - rating_numerator : the rating at the top
  - retweet_count : numebers of retweets
  - favorite_count: numbers of favorites
  - j_url: the link of tweet images 
  - img_num : the image number betweebn 1- 4
  - confidence : the confidence level of the image after running it through a neural network
